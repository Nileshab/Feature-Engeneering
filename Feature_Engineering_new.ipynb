```python
data = pd.read_csv("/content/denmark - denmark.csv")
data.shape
```

```python
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler().set_output(transform = "pandas")
data_sc = sc.fit_transform(data)
data_sc.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
mnsc = MinMaxScaler().set_output(transform = "pandas")
data_mnsc = mnsc.fit_transform(data)
data_mnsc.head()
```

```python
q3 = data.iloc[:,1].quantile(0.75)
q1 = data.iloc[:,1].quantile(0.25)
iqr = q3-q1
upper = q3+1.5*iqr
lower = q1-1.5*iqr
data.iloc[:,1] = data.iloc[:,1].apply(lambda x: upper if x>upper else x ).apply(lambda x: lower if x<lower else x)
```

```python
for i in range(len(data_mnsc.columns)):
  q3 = data_mnsc.iloc[:,i].quantile(0.75)
  q1 = data_mnsc.iloc[:,i].quantile(0.25)
  iqr = q3-q1
  upper = q3+1.5*iqr
  lower = q1-1.5*iqr
  data_mnsc.iloc[:,i] = data_mnsc.iloc[:,i].apply(lambda x: upper if x>upper else x ).apply(lambda x: lower if x<lower else x)
```

```python
import pandas as pd
import numpy as np

# Set a random seed for reproducibility
np.random.seed(42)

# Define the number of rows
n_rows = 10

# Generate numerical data (e.g., age, income)
age = np.random.randint(18, 60, size=n_rows)
income = np.random.randint(30000, 100000, size=n_rows)

# Generate categorical data (e.g., gender, city)
gender = np.random.choice(['Male', 'Female'], size=n_rows)
city = np.random.choice(['New York', 'Los Angeles', 'Chicago', 'Houston'], size=n_rows)

# Create a DataFrame
data = {
    'Age': age,
    'Income': income,
    'Gender': gender,
    'City': city
}

df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
le = LabelEncoder()
df["Gender"] = le.fit_transform(df["Gender"])
df["City"] = le.fit_transform(df["City"])
```
